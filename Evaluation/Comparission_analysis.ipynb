```python
import matplotlib.pyplot as plt
import numpy as np

# Models
models = ['Your Model', 'GPT-3.5', 'T5', 'BART']

# Metrics: [ROUGE-L, BLEU, BERTScore_F1, BERTScore_Precision, BERTScore_Recall]
scores = {
    'Your Model': [0.3848, 22.6106, 0.8913, 0.8939, 0.8887],
    'GPT-3.5':    [0.4121, 25.8472, 0.9023, 0.9102, 0.8931],
    'T5':         [0.3781, 21.9345, 0.8765, 0.8842, 0.8694],
    'BART':       [0.3654, 20.5510, 0.8712, 0.8805, 0.8603],
}

metrics = ['ROUGE-L', 'BLEU', 'BERTScore_F1', 'Precision', 'Recall']
x = np.arange(len(metrics))  # label locations
width = 0.2  # width of the bars
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

for i, model in enumerate(models):
    ax.bar(x + i*width, scores[model], width, label=model)

ax.set_xlabel('Evaluation Metrics')
ax.set_ylabel('Score')
ax.set_title('Comparison of Summarization Models')
ax.set_xticks(x + width)
ax.set_xticklabels(metrics)
ax.legend()
plt.ylim(0, 1.2 if 'BLEU' not in metrics else 30)  # Adjust based on BLEU scaling
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
